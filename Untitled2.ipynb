```python
hop_seconds = 1
window_seconds = 2
n_mfcc = 20
```

```python
mfcc=librosa.feature.mfcc(y=y, sr=sr,
                          hop_length=int(hop_seconds*sr),
                          n_fft=int(window_seconds*sr),
                          n_mfcc=n_mfcc)
mfcc_delta=librosa.feature.delta(mfcc)
mfcc_delta2=librosa.feature.delta(mfcc, order=2)
stacked=np.vstack((mfcc, mfcc_delta, mfcc_delta2))
features=stacked.T #librosa возвращает где MFCC идут в ряд, а для модели нужно будет в столбец.
```

```python
# Feature Generation
# MFCCs
mfcc_mean=np.mean(mfcc,axis=1)
mfcc_std=np.std(mfcc,axis=1)

coeffs=np.arange(0,20)
plt.figure(figsize=(15,5))
plt.title('Mean MFCCs')
sns.barplot(x=coeffs,y=mfcc_mean)

plt.figure(figsize=(15,5))
plt.title('SD MFCCs')
sns.barplot(x=coeffs,y=mfcc_std)
# Generate the chroma Dataframe
mfcc_df=pd.DataFrame()
for i in range(0,20):
    mfcc_df['mfccs_mean_'+str(i)]=mfcc_mean[i]
for i in range(0,20):
    mfcc_df['mfccs_std_'+str(i)]=mfcc_mean[i]
mfcc_df.loc[0]=np.concatenate((mfcc_mean,mfcc_std),axis=0)
mfcc_df
```
